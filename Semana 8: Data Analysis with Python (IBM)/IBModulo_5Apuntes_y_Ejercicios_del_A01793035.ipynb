Seleccionamos el valor de alfa que minimiza el error de prueba. Para hacerlo, podemos usar un bucle for. Tambin hemos creado una barra de progreso para ver cuntas iteraciones hemos completado hasta ahora.

```python
from tqdm import tqdm

Rsqu_test = []
Rsqu_train = []
dummy1 = []
Alpha = 10 * np.array(range(0,1000))
pbar = tqdm(Alpha)

for alpha in pbar:
    RigeModel = Ridge(alpha=alpha) 
    RigeModel.fit(x_train_pr, y_train)
    test_score, train_score = RigeModel.score(x_test_pr, y_test), RigeModel.score(x_train_pr, y_train)
    
    pbar.set_postfix({"Test Score": test_score, "Train Score": train_score})

    Rsqu_test.append(test_score)
    Rsqu_train.append(train_score)
```

Podemos trazar el valor de R^2 para diferentes alfas:

```python
width = 12
height = 10
plt.figure(figsize=(width, height))

plt.plot(Alpha,Rsqu_test, label='validation data  ')
plt.plot(Alpha,Rsqu_train, 'r', label='training Data ')
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.legend()
```

La lnea azul representa el R^2 de los datos de validacin y la lnea roja representa el R^2 de los datos de entrenamiento. El eje x representa los diferentes valores de Alpha.

Aqu, el modelo se construye y prueba con los mismos datos, por lo que los datos de entrenamiento y prueba son los mismos.

La lnea roja en la Figura 4 representa el R^2 de los datos de entrenamiento. A medida que alfa aumenta, R ^ 2 disminuye. Por lo tanto, a medida que aumenta alfa, el modelo funciona peor en los datos de entrenamiento

La lnea azul representa el R^2 en los datos de validacin. A medida que aumenta el valor de alfa, R ^ 2 aumenta y converge en un punto.
